```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load custom dataset
data = pd.read_csv('data.csv')

# Set up the matplotlib figure
plt.figure(figsize=(14, 10))

# 1. Scatter Plot (e.g., Age vs. Score, no hue as there's no categorical variable)
plt.subplot(231)  # 2x3 grid, 1st subplot
sns.scatterplot(x='Age', y='Score', data=data, color='blue')
plt.title('Scatter Plot: Age vs Score')

# 2. Histogram (e.g., Age distribution)
plt.subplot(232)  # 2x3 grid, 2nd subplot
sns.histplot(data['Age'], kde=True, color='green', bins=10)
plt.title('Histogram of Age')

# 3. Bar Plot (e.g., Average Score by Name)
plt.subplot(233)  # 2x3 grid, 3rd subplot
sns.barplot(x='Name', y='Score', data=data, palette='muted')
plt.title('Bar Plot: Average Score by Name')

# 4. Box Plot (e.g., Score distribution by Name)
plt.subplot(234)  # 2x3 grid, 4th subplot
sns.boxplot(x='Name', y='Score', data=data, palette='Set3')
plt.title('Box Plot: Score by Name')

# 5. Violin Plot (e.g., Score distribution by Name)
plt.subplot(235)  # 2x3 grid, 5th subplot
sns.violinplot(x='Name', y='Score', data=data, palette='coolwarm')
plt.title('Violin Plot: Score by Name')

# 6. Pair Plot (Not meaningful here since there are only 3 numerical columns)
plt.subplot(236)  # 2x3 grid, 6th subplot
sns.pairplot(data, diag_kind='kde', palette='dark', markers=["o", "s", "D"])
plt.suptitle('Pair Plot: Correlations between Variables', y=1.02)

# Adjust layout to prevent overlapping of subplots
plt.tight_layout()
plt.show()
```
